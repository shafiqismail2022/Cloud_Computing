## Importing necessary libraries

```python
import pandas as pd #importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

## Exploratory Data Analysis

```python
plt.figure(figsize=(10,6)) #figure size
sns.heatmap(titanic.isnull(),cmap='plasma',cbar=False,yticklabels=False) #heatmap showing null values
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median()) # Handling the null values for age with mean 
titanic['cabin'] = titanic['cabin'].fillna('Unknown')# Handling the null values for the Cabin
titanic = titanic.drop(['boat', 'body','home.dest'], axis=1) # we removed some data that had alot of null values
```

```python
rows_with_null = titanic[titanic.isnull().any(axis=1)] # finding null data
print(rows_with_null) # printing the same
```

```python
mode_value = titanic['fare'].mode()[0] # finding the value of mode
print(mode_value)
titanic['fare'].fillna(mode_value, inplace=True) # filling the missing value of fare with mode
```

```python
plt.figure(figsize=(10,6)) 
sns.heatmap(titanic.isnull(),cmap='plasma',cbar=False,yticklabels=False) #heatmap to check if they were missing values
```

```python
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter') #ploting a joint plot of fare and age
```

```python
sns.swarmplot(x='pclass',y='age',data=titanic,palette='viridis') # visualizing the relationship between pclass and age
```

```python
sns.countplot(x='sex',data=titanic,palette='viridis',hue='survived') #visualizing the survived and dead based gender
```

```python
grid2=sns.FacetGrid(data=titanic,col='sex',hue='survived') 
grid2.map(plt.hist,'age') # visualizing survived and dead with gender and age 
plt.legend()
```

## Models 

### Logistic Regression Model

```python
titanic.head() # checking the data
```

```python
titanic = pd.get_dummies(titanic, columns=['sex', 'embarked'], drop_first=True) # creating dummies value for categorical values
```

```python
titanic.head() # checking the new head data
```

```python
titanic.columns # printing columns of the data
```

```python
features=['pclass', 'age', 'sibsp', 'parch', 'fare', 'sex_male', 'embarked_Q', 'embarked_S'] #listing the required features for modeling
```

```python
from sklearn.linear_model import LogisticRegression # importing logistic regression model
from sklearn.model_selection import train_test_split # importing train and test data split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report # importing evaluation matrics
```

```python
from sklearn.preprocessing import StandardScaler # importing Z-score Normalization model 
```

```python
X=titanic[features] # input data
y=titanic['survived'] # label data
```

```python
scaler = StandardScaler() # instantiating a standard scale to perform Z-score normalization
X_scaled = scaler.fit_transform(X) # training and transforming the data
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.33,random_state=101) #spliting the data into train and test set
```

```python
lg=LogisticRegression(max_iter=1000) #instantiating logistic Regression
```

```python
lg.fit(X_train,y_train) #training the data
```

```python
lg_pred=lg.predict(X_test) # predicting for the test set
```

```python
accuracy_lg = accuracy_score(y_test, lg_pred) # accuracy for logistic model
conf_matrix_lg = confusion_matrix(y_test, lg_pred) # confusion matrix
class_report_lg = classification_report(y_test, lg_pred) #classification  report metrics
```

```python
print(f"Accuracy: {accuracy_lg:.3f}") #printing the metrics
print("Confusion Matrix:")
print(conf_matrix_lg)
print("Classification Report:")
print(class_report_lg)
```

### Decision Tree Classifier 

```python
from sklearn.tree import DecisionTreeClassifier # importing decision tree classifier
```

```python
dt3 = DecisionTreeClassifier(max_depth=4,splitter='best',criterion='entropy',random_state=101,) # instantiating the decision tree classifier
```

```python
dt3.fit(X_train,y_train) #training the data
```

```python
dt3_pred=dt3.predict(X_test) # predicting the data
```

```python
accuracy_dt3 = accuracy_score(y_test, dt3_pred) # metrics for
conf_matrix_dt3 = confusion_matrix(y_test, dt3_pred) # confusion metrics
class_report_dt3 = classification_report(y_test, dt3_pred) # classification report metrics
```

```python
print(f"Accuracy: {accuracy_dt3:.3f}") # printing metrics
print("Confusion Matrix:")
print(conf_matrix_dt3)
print("Classification Report:")
print(class_report_dt3)
```

### Support Vector Machine

```python
from sklearn.svm import SVC # importing Support vector machine
from sklearn.decomposition import PCA # importing principal component analysis
```

```python
model=SVC(random_state=101) # instantiating the SVM
```

```python
pca = PCA(n_components=2)   # instantiating principal component analysis
X_train_pca = pca.fit_transform(X_train) #  train and fit the data for PCA
X_test_pca = pca.transform(X_test) # transforming data for PCA
```

```python
model.fit(X_train_pca,y_train) # training the data
```

```python
svm_pred=model.predict(X_test_pca) #predicting the data
```

```python
accuracy_svm=accuracy_score(y_test, svm_pred) # accuracy metric
conf_matrix_svm=confusion_matrix(y_test, svm_pred) # confusion metric
class_report_svm=classification_report(y_test, svm_pred) # classification report metrics
```

```python
print(f"Accuracy: {accuracy_svm:.3f}") # printing metrics
print("Confusion Matrix:")
print(conf_matrix_svm)
print("Classification Report:")
print(class_report_svm)
```

### K-Mean(Unsuperised Learning Algorithm)

```python
from sklearn.cluster import KMeans # importing the kmean clustering
```

```python
km=KMeans(n_clusters=2) # instantiating kmean clustering
```

```python
km.fit(X_train_pca) #train the data
```

```python
km.labels_ # get the labels
```

```python
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=km.labels_, cmap='viridis', marker='o') # ploting the scatter graph between the principal component analysis and cluster labels
plt.title('KMeans Clustering with PCA') # title of the graph
plt.xlabel('Principal Component 1') # label for x-axis  
plt.ylabel('Principal Component 2') # label for y-axis
plt.show()
```
